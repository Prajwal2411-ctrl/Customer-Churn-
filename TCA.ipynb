```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"

df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
    
```

#converted 0 and 1 values of senior citizen to yes/no to make it easier to understand 

```python
ax = sns.countplot(x = 'Churn', data = df)

ax.bar_label(ax.containers[0])
plt.title('count of customers by churn')
plt.show()
```

```python
plt.figure(figsize = (3,4))
gb = df.groupby("Churn").agg({'Churn':"count"})
plt.pie(gb['Churn'],labels = gb.index, autopct = '%1.2f%%')
plt.title('Percentage Of Churned Customers', fontsize = 10)
plt.show()
```

```python

# List of columns to plot
columns = [
    'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 
    'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Determine the number of rows and columns for the subplots (3x3 grid for 9 columns)
n_cols = 3
n_rows = (len(columns) + n_cols - 1) // n_cols  # Calculate required rows for a grid

# Set up the matplotlib figure
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 10))
fig.suptitle('Count Plots of Different Services', fontsize=16)

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Create each count plot in a subplot
for i, col in enumerate(columns):
    sns.countplot(x=col, data=df, ax=axes[i], hue = ["Churn"])
    axes[i].set_title(col)
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')

# Remove empty subplots if columns list length is not a perfect grid fit
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout(rect=[0, 0, 1, 0.96])  # Adjust layout to fit the title
plt.show()
```

#Each plot illustrates that customers with certain services are more likely to churn, such as those with Fiber optic internet. In contrast, services like PhoneService and MultipleLines have fewer differences in churn rates between customers. These insights can help identify which services are associated with higher churn.
